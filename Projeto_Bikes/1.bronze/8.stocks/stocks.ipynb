```python
# Definição de paths
resource_path = 'abfss://bikestore@azurelakehouse.dfs.core.windows.net/resource/origem'
bronze_path = 'abfss://bikestore@azurelakehouse.dfs.core.windows.net/bronze/'
```

```python
df_stocks = spark.read.csv(f'{resource_path}/stocks.csv',
    header=True,
    inferSchema=True,
    sep=','
)
```

```python
df_stocks.write\
    .mode('overwrite')\
    .format('delta')\
    .save(f"{bronze_path}stocks")
```
